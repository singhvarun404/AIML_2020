```python
### Drawing different shapes on image
```

```python
### Drawing a line
img = np.zeros((512,512,3),np.uint8)+255 ### Blank white image
cv2.line(img,(55,100),(400,500),(0,0,255),6) 
### line(image, 1st point, 2nd point, color, width)

plt.imshow(img)
plt.show()
```

```python
### Drawing a rectangle
img = np.zeros((512,512,3),np.uint8)+255 ### Blank white image
cv2.rectangle(img,(55,100),(400,500),(0,0,255),-1)
### rectangle(image, diagnal corner pnt, diagnal corner pnt,color, width) if width = -1 then rectangle will be filled

plt.imshow(img)                                     
plt.show()                                         
```

```python
### Drawing a Circle
img = np.zeros((512,512,3),np.uint8)+255 ### Blank white image
cv2.circle(img,(250,250),150,(160,140,255),4)
### Circle(image, centre, radius, color, width)

cv2.circle(img,(250,250),5,(0,220,120),-1)        
cv2.circle(img,(250,250),7,(0,0,0),2)           
plt.imshow(img)                                    
plt.show()                                         
```

```python
### Drawing a polygon
img = np.zeros((512,512,3),np.uint8)+255 ### Blank white image
points = np.array([[10,100],[190,300],[390,100]]) 
### points of polygon

cv2.polylines(img,[points],True,[0,0,255],5)
### polylines(image,[points],isclosed,color,width)

plt.imshow(img)                                    
plt.show()                                         
```

```python
### Writting text on image
img = np.zeros((512,512,3),np.uint8)+255 ### Blank white image
cv2.putText(img,"Hello World!",(150,250),cv2.FONT_HERSHEY_SCRIPT_COMPLEX,1.5,(0,0,255),1) 
### putText(image,text,point,style,size,color,width)

plt.imshow(img)                                    
plt.show()    
```
